```python
import pandas as pd
```

```python
df1 = pd.read_csv("Resources/youtube_videos.csv", encoding='ISO-8859-1')
df2 = pd.read_csv("Resources/youtube_videos (1).csv", encoding='ISO-8859-1')
df3 = pd.read_csv("Resources/youtube_videos3.csv", encoding='ISO-8859-1')
```

```python
# Merge df1, df2, and df3 separately
merged_df1 = pd.merge(df1, df2, how='outer')
merged_df2 = pd.merge(merged_df1, df3, how='outer')

# Concatenate and drop duplicates based on 'Video Name' column
merged_df = pd.concat([df1, df2, df3]).drop_duplicates(subset='Video Name')

# Export the final merged dataframe to a new CSV file
merged_df.to_csv('/Users/issiliasette/Desktop/data_course/project1/Porject/Resources/merged_videos.csv', index=False)

```
